```python
#Reading Members Data
for k,v in info.items():
    sender = k
    receiver = k
    password = v
    
    for i in range(0,15):
        maindata = pd.read_excel(collegedata)
        maindata.rename(columns={maindata.columns[0]: "Email"}, inplace=True)
        data = maindata.head(90)
        print(sender + " : " + password)
        print("Sending mails to " + str(data.shape[0]) + " ids")
        
        if(data.shape[0] > 1):
            print("top row in set: " + str(maindata["Email"][0]))
            print("top row in subset: " + str(data["Email"][0]))
            print("last row in set: " + str(maindata["Email"][data.shape[0] - 1]))
            print("last row in subset: " + str(data["Email"][data.shape[0] - 1]))
            
            del message["From"]
            del message["To"]
            message["From"] = sender
            message["To"] = receiver

            #Sending Mails
            try:
                session = smtplib.SMTP("smtp.gmail.com", 587)
                session.starttls()
                session.login(sender, password)
                text = message.as_string()
                session.sendmail(sender, [receiver]+data["Email"].to_list(), text)
                session.quit()
            except Exception as e:
                print("##" + str(e))
                raise ValueError(str(e))

            maindata = maindata.iloc[data.shape[0]:]
            maindata = maindata.reset_index(drop=True)
            maindata.to_excel(collegedata, index = False)
            print("##################\n")
        else:
            print("No more ids to select from\n##################\n")
```
